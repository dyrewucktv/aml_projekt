```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(rc={'figure.figsize':(8,6)})
```

```python
df = pd.read_csv('classification_performance.csv')
```

```python
df
```

```python
sns.set_theme(style="whitegrid")
sns.boxplot(data=df[df["product"] == True], x="dataset", y="accuracy", hue="method")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title("Classification performance on product datasets")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
sns.boxplot(data=df[df["product"] == False], x="dataset", y="accuracy", hue="method")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xticks(rotation=45)
# set title
plt.title("Classification accuracy for non-product datasets")
plt.show()
```

```python
df_true = df[df["product"] == True]
```

```python
df_true.groupby(["dataset", "method"]).agg({"accuracy": [("mean_accuracy", "mean"), ("std_accuracy", "std")]}).sort_values(by=["dataset", ("accuracy", "mean_accuracy")], ascending=[True, False])
```

```python
df_false = df[df["product"] == False]
```

```python
# display all rows
pd.set_option('display.max_rows', None)

df_false.groupby(["dataset", "method"]).agg({"accuracy": [("mean_accuracy", "mean"), ("std_accuracy", "std")]}).sort_values(by=["dataset", ("accuracy", "mean_accuracy")], ascending=[True, False])
```
